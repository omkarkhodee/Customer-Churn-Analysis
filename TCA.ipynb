```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a crosstab of counts
counts = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Convert counts to percentages
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Step 3: Plot as stacked bar chart
ax = percentages.plot(kind='bar', stacked=True, figsize=(3, 3), colormap='Set2')

# Step 4: Add percentage labels
for i, (index, row) in enumerate(percentages.iterrows()):
    cumulative = 0
    for col in percentages.columns:
        height = row[col]
        if height > 0:
            ax.text(i, cumulative + height / 2, f'{height:.1f}%', ha='center', va='center', fontsize=8, color='black')
        cumulative += height

# Formatting
plt.title("Churn by SeniorCitizen")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")
plt.legend(title="Churn", loc="upper right")
plt.tight_layout()
plt.show()
```

comparitive a high percentage people in senior citizen category have churned

```python


# Features to plot
features = ['PhoneService', 'MultipleLines', 'InternetService',
            'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
            'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up subplot grid (3 columns)
n_cols = 3
n_rows = -(-len(features) // n_cols)  # Ceiling division

# Set the figure size
plt.figure(figsize=(n_cols * 5, n_rows * 4))

# Plot each feature
for i, feature in enumerate(features, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.countplot(data=df, x=feature, hue='Churn', palette='Set2')
    plt.title(f'Churn by {feature}', fontsize=12)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.legend(title='Churn', loc='upper right')

# Adjust layout
plt.tight_layout()
plt.show()
```

Customers with no internet-related services (like OnlineSecurity, TechSupport, or Streaming) show lower churn, likely due to them not being active users.
Services like Fiber optic Internet and MultipleLines have noticeably higher churn rates compared to DSL or single-line users.
OnlineSecurity and TechSupport users tend to churn less, suggesting these services may help with customer retention.
